```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("titanic.csv")
df
```

```python
df=df.replace('?',None)
```

```python
df
```

```python
df.info()
```

## since cabin, boat, body have significant absence of details we drop those columns

```python
df=df.drop(["cabin", "boat","body"], axis=1)
```

```python
df
```

```python
df.dropna(axis=0)
```

```python
df.survived.replace([1,0],["Yes","No"], inplace=True)
```

```python
tab = pd.crosstab(df['sex'], df['survived'])
print(tab)
```

```python
result = tab.div(tab.sum(1).astype(float), axis=0).plot(kind="bar", stacked=True)
plt.xlabel('Port embarked')
plt.ylabel('Percentage')
```

```python
plt.figure
sns.countplot(data=df,x=df.pclass)
```

```python
plt.figure
survived_class=df[(df.survived=="Yes")]
sns.countplot(data=survived_class,x=survived_class.pclass)
```

```python
sns.violinplot(x = 'embarked', y = 'pclass', data = df, hue = 'sex')
```
